```python
import pandas as pd
from collections import Counter # Import Counter

#Load the dataset
df = pd.read_csv('/content/netflix_titles.csv')

#Preview the data
df.head()

#Check the shape of the dataset
df.shape

#See column names and types
df.info()

#Check for missing data
df.isnull().sum()

#How many Movies vs TV Shows
df['type'].value_counts()

#Find top 10 countries with the most Netflix content
df['country'].value_counts().head(10)

#Count how many titles were released each year
df['release_year'].value_counts().sort_index(ascending=True).plot(kind='bar', figsize=(12, 6), title='Netflix Titles by Release Year')

#What are te most common genres?
df['listed_in'].value_counts().head(10)

#Who appears most often in Netflix titles?

#Drop rows with no cast listed
df_cast = df.dropna(subset=['cast'])

#Split cast names and flatten the list
all_cast = ','.join(df_cast['cast']).split(',')
cast_counts = Counter([name.strip() for name in all_cast])
cast_counts.most_common(10)

#Movies vs TV Shows Pie Chart
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#66b3ff','#ff9999'], figsize=(6,6), title='Distribution of Content Types')
plt.ylabel('')
plt.show()

#Top 10 Countries With Most Content (Bar Chart)
df['country'].value_counts().head(10).plot(kind='bar', color='purple', title='Top 10 Countries with Netflix Content', figsize=(10,5))
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```
